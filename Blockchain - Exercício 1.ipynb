```python
#Importando bibliotecas que seram utlizadas
import hashlib
import datetime
import time
```

```python
#Função que transforma String em Hash256
def hash(texto):
    md5_hash = hashlib.sha256((texto).encode('utf-8'))
    
    return md5_hash.hexdigest()
```

```python
#Criando uma lista vazia
blocos = []
```

```python
#Função para criar bloco Genêsis
def genesis():
    cont = 1
    data_local = datetime.date.today()
    
    conteudo = {
        "index":"Bloco " + str(len(blocos)),
        "timestamp": data_local,
        "msg": "Heloo World!"
    }
    
    conteudo['nonce'] = cont
    
    transform = conteudo['index'] + str(conteudo['timestamp']) + str(conteudo['msg']) + str(conteudo['nonce'])
    
    hash256 = hash(transform)
    
    while hash256[0:4] != '0000':
        
        if hash256[0:4] != '0000':
            cont += 1

            conteudo['nonce'] = cont

            transform = conteudo['index'] + str(conteudo['timestamp']) + str(conteudo['msg']) + str(conteudo['nonce'])

            hash256 = hash(transform)

        conteudo['hash'] = hash256

    blocos.append(conteudo)
    print("####################BLOCO GENESIS####################")
    ##print(f"{conteudo}")
    for key, value in conteudo.items():
        print(f'{key}: {value}')
```

```python
#Função para criar restante da caceia de blocos
def cadeia(qtd):
    while len(blocos) < qtd:
        cont = 1
        data_local = datetime.date.today()

        conteudo = {
            "index":"Bloco " + str(len(blocos) - 1),
            "timestamp": data_local,
            "msg": "Heloo World!",
            "hash_anterior": blocos[-1]['hash']
        }

        conteudo['nonce'] = cont

        transform = conteudo['index'] + str(conteudo['timestamp']) + str(conteudo['msg']) + conteudo['hash_anterior'] + str(conteudo['nonce'])

        hash256 = hash(transform)

        while hash256[0:4] != '0000':

            if hash256[0:4] != '0000':
                cont += 1

                conteudo['nonce'] = cont

                transform = conteudo['index'] + str(conteudo['timestamp']) + str(conteudo['msg']) + conteudo['hash_anterior'] + str(conteudo['nonce'])

                hash256 = hash(transform)

            conteudo['hash'] = hash256

        blocos.append(conteudo)
        print(f"####################BLOCO {len(blocos)}####################")
        for key, value in conteudo.items():
            print(f'{key}: {value}')
```

```python
#Função para converter segundos em tempo, para saber tempo de execução do job
def convert(seconds): 
    seconds = seconds % (24 * 3600) 
    hour = seconds // 3600
    seconds %= 3600
    minutes = seconds // 60
    seconds %= 60
      
    return "%d:%02d:%02d" % (hour, minutes, seconds)
```

```python
#Função principal roda as duas funções de criação de blocos
def main(qtd):
    #inicio da execução do job
    start = time.time()
    
    genesis()
    cadeia(qtd)
    
    #fim da execução do job
    end = time.time()
    dif = end - start
    tempo = convert(dif)
    
    print(f"Tempo de execução do job {tempo}")
```

```python
main(100)
```
